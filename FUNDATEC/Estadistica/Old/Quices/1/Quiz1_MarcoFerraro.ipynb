```python
means = [1.230, 1.330, 0.040, 0.044, 1.200, 0.270, 0.490, 0.190, 0.830, 0.810, 0.710, 0.560, 0.490, 1.160, 0.050, 0.150, 0.195, 0.770, 1.080, 0.980, 0.630, 0.560, 0.410, 0.730, 0.590, 0.340, 0.340, 0.840, 0.500, 0.3400, 0.280, 0.340,  0.750, 0.870, 0.560, 0.170, 0.180, 0.190, 0.040, 0.490, 1.100, 0.160, 0.100, 0.2100, 0.860, 0.520, 0.650, 0.270, 0.940, 0.400, 0.460, 0.260, 0.270]

len(means)
```

```python
import torch

def calculate_mean(data_list):
    tensor_data = torch.tensor(data_list)
    mean = torch.mean(tensor_data)

    return mean
```

```python
def calculate_std(data_list):
    data = torch.tensor(data_list)
    std= data.std(unbiased=False)

    return std
```

```python
mean = calculate_mean(means)
mean
```

```python
std = calculate_std(means)
std
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_normal_distribution(mu, sigma, num_sigma=3):
    x = np.linspace(mu - (num_sigma * (sigma)), mu + (num_sigma * (sigma)), 1000)
    pdf = norm.pdf(x, mu, sigma)

    plt.plot(x, pdf, color='b')

    lower_bound = mu - (num_sigma * sigma)
    upper_bound = mu + (num_sigma * sigma)

    plt.fill_between(x, pdf, where=(x >= lower_bound) & (x <= upper_bound), color='yellow', alpha=0.3)

    lower_bound = round(lower_bound, 3)
    upper_bound = round(upper_bound, 3)

    plt.axvline(x=lower_bound, color='red', linestyle='--', label=f'Lower Bound ({lower_bound:.3f} Sigma)')
    plt.axvline(x=upper_bound, color='green', linestyle='--', label=f'Upper Bound ({upper_bound:.3f} Sigma)')

    plt.xlabel('Valores')
    plt.ylabel('Función de Densidad de Probabilidad')
    plt.title(f'Función de Densidad Normal (μ={mu:.3f}, σ={sigma:.3f})')
    plt.legend(loc='upper right')

    plt.grid(True)
    plt.show()
```

```python
mu_np = mean.numpy()
sigma_np = std.numpy()

plot_normal_distribution(mu=mu_np, sigma=sigma_np, num_sigma=3)
```
