#### 1.2 Estadística Descriptiva
Se realiza un análisis estadístico de los datos, utilizando la función `get_statistics`. Esta función calcula la media, mediana, desviación estándar, mínimo, máximo entre otras variables estadísticas. Se utiliza la librería Pandas para realizar este análisis.

```python
def get_statistics(df):

    
    result = {
        'mean': df.mean(),
        'median': df.median(),
        'mode': df.mode()[0],
        'variance': df.var(),
        'standard deviation': df.std(),
        'max': df.max(),
        'min': df.min(),
        'range': df.max() - df.min(),
        'quantile 1': df.quantile(0.25),
        'quantile 2': df.quantile(0.5),
        'quantile 3': df.quantile(0.75),
        'interquartile range': df.quantile(0.75) - df.quantile(0.25),
        'skewness': df.skew(),
    }
    
    
    result = pd.DataFrame(result, index=[0])
    return result
```

A continuación se muestran los resultados obtenidos de varias variables del dataset, como `Rings`, `Length`, `Diameter`, `Height`, `Shell Weight`, `Whole`, `Viscera`. Cabe destacar que la variable `Rings` es la variable objetivo, y las demás son variables predictoras. Además es importante mencionar que este método no es aplicable a variables categóricas, por lo que no se incluyen en este análisis.

```python
get_statistics(df['Rings'])
```

```python
get_statistics(df['Length'])
```

```python
get_statistics(df['Diameter'])
```

```python
get_statistics(df['Height'])
```

```python
get_statistics(df['Shellweight'])
```

```python
get_statistics(df['Whole'])
```

#### 1.3.3 Hexbin

Se realiza un gráfico de densidad entre las variables `Diameter` y `Height`. Se utiliza la librería Matplotlib para realizar este análisis. La idea es comprender la relación entre estas dos variables.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap

def plot_hex_bin(x, y, x_label, y_label, title):
    colors = [(0.0, 'white'), (0.5, 'blue'), (1.0, 'red')]
    cmap = LinearSegmentedColormap.from_list('custom_cmap', colors)

    # Create the hexagonal binning plot
    plt.hexbin(x, y, gridsize=20, cmap=cmap)

    # Add a colorbar
    plt.colorbar()

    # Add labels to the x-axis and y-axis
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(title)

    # Display the plot
    plt.show()
```

```python
plot_hex_bin(df['Height'], df['Diameter'], 'Abalone Height', 'Abalone Diameter', 'Hexagonal Binning of Abalone Height vs Diameter')
```

Al ver los gráficos de densidad, podemos apreciar que la relación entre las variables `Diameter` y `Height` es relativamente lineal. Esto nos indica que estas dos variables tienen una relación lineal positiva. Ademas, debido a la naturaleza del grafico, podemos ver los puntos donde los presenta una mayor densidad de datos. Es precisamente, en los puntos de color rojo donde se concentran la mayor cantidad de datos.

```python
plot_hex_bin(df['Height'], df['Shellweight'], 'Abalone Height', 'Abalone Shellweight', 'Hexagonal Binning of Abalone Height vs Shellweight')
```

Vemos que al analizar `Height` y `Shellwight` se obtiene un gráfico de densidad similar al anterior. Esto nos indica que estas dos variables también tienen una relación lineal positiva pero no tan demarcada como la anterior.

```python
plot_hex_bin(df['Diameter'], df['Shellweight'], 'Abalone Diameter', 'Abalone Shellweight', 'Hexagonal Binning of Abalone Diameter vs Shellweight')
```

Al analizar `Diameter` y `Shellweight` ocurre algo similar.
