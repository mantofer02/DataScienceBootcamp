#### 1.2 Estadística Descriptiva
Se realiza un análisis estadístico de los datos, utilizando la función `get_statistics`. Esta función calcula la media, mediana, desviación estándar, mínimo, máximo entre otras variables estadísticas. Se utiliza la librería Pandas para realizar este análisis.

```python
def get_statistics(df):

    
    result = {
        'mean': df.mean(),
        'median': df.median(),
        'mode': df.mode()[0],
        'variance': df.var(),
        'standard deviation': df.std(),
        'max': df.max(),
        'min': df.min(),
        'range': df.max() - df.min(),
        'quantile 1': df.quantile(0.25),
        'quantile 2': df.quantile(0.5),
        'quantile 3': df.quantile(0.75),
        'interquartile range': df.quantile(0.75) - df.quantile(0.25),
        'skewness': df.skew(),
    }
    
    
    result = pd.DataFrame(result, index=[0])
    return result
```

A continuación se muestran los resultados obtenidos de varias variables del dataset, como `Rings`, `Length`, `Diameter`, `Height`, `Shell Weight`, `Whole`, `Viscera`. Cabe destacar que la variable `Rings` es la variable objetivo, y las demás son variables predictoras. Además es importante mencionar que este método no es aplicable a variables categóricas, por lo que no se incluyen en este análisis.

```python
get_statistics(df['Rings'])
```

#### 1.3.3 Hexbin

Se realiza un gráfico de densidad entre las variables `Diameter` y `Height`. Se utiliza la librería Matplotlib para realizar este análisis. La idea es comprender la relación entre estas dos variables.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap

def plot_hex_bin(x, y, x_label, y_label, title):
    colors = [(0.0, 'white'), (0.5, 'blue'), (1.0, 'red')]
    cmap = LinearSegmentedColormap.from_list('custom_cmap', colors)

    # Create the hexagonal binning plot
    plt.hexbin(x, y, gridsize=20, cmap=cmap)

    # Add a colorbar
    plt.colorbar()

    # Add labels to the x-axis and y-axis
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(title)

    # Display the plot
    plt.show()
```

```python
plot_hex_bin(df['Height'], df['Diameter'], 'Abalone Height', 'Abalone Diameter', 'Hexagonal Binning of Abalone Height vs Diameter')
```

Al ver los gráficos de densidad, podemos apreciar que la relación entre las variables `Diameter` y `Height` es relativamente lineal. Esto nos indica que estas dos variables tienen una relación lineal positiva. Ademas, debido a la naturaleza del grafico, podemos ver los puntos donde los presenta una mayor densidad de datos. Es precisamente, en los puntos de color rojo donde se concentran la mayor cantidad de datos.

```python
plot_hex_bin(df['Height'], df['Shellweight'], 'Abalone Height', 'Abalone Shellweight', 'Hexagonal Binning of Abalone Height vs Shellweight')
```

### 4. Remplazo de Outliers

Los outliers son valores que se encuentran muy alejados de la distribución de los datos. Estos valores pueden afectar el análisis de los datos, por lo que se procede a removerlos. Como se ha mencionado anteriormente, la columna `Rings` presenta outliers. Para remover estos outliers el primer paso es detectarlos. Para esto se va a utilizar un algoritmo llamado Local Outlier Factor (LOF). Este algoritmo detecta outliers utilizando el concepto de densidad local. Este algoritmo se encuentra implementado en la librería Scikit-Learn.

Al aplicar el algoritmo LOF se obtiene un arreglo de factores. Si el valor es 1 significa que el dato tiene un alto grado de probabilidad de no ser un outlier. Si el valor es menor a 1 significa que el dato tiene un alto grado de probabilidad de ser un outlier. 

Para respetar la naturaleza de los datos, se remplazara a los outliers con el valor más cercano que no sea un outlier. 

```python
import numpy as np
from sklearn.neighbors import LocalOutlierFactor

X = df['Rings'].to_numpy()
X = X.reshape(-1, 1)

clf = LocalOutlierFactor()
outlier_labels = clf.fit_predict(X)

outlier_indices = np.where(outlier_labels == -1)[0]
non_outlier_indices = np.where(outlier_labels == 1)[0]

outlier_labels
```

Para calcular el outlier más cercano obtenemos la diferencia absoluta o distancia entre el outlier y todos los valores que no son outliers. Luego, se obtiene el valor mínimo de esta diferencia. Finalmente, se remplaza el outlier con el valor que tiene la diferencia mínima.

```python

for outlier_index in outlier_indices:
    distances = np.abs(X[outlier_index] - X[non_outlier_indices])
    closest_non_outlier_index = non_outlier_indices[np.argmin(distances)]
    df.loc[outlier_index, 'Rings'] = df.loc[closest_non_outlier_index, 'Rings']

df['Rings']
```

Volvemos a graficar el boxplot de la variable `Rings` sin los outliers. Podemos ver que la distribución de los datos es mucho más uniforme.

```python
plt.boxplot(df['Rings'], vert=False)

plt.xlabel('Amount of Rings')
plt.ylabel('Boxplot')
plt.title('Boxplot of Rings in Abalone Dataset')

plt.show()
```

### 5. Remplazo de valores nominales

La columna `Sex` del df posee valores nominales. Se utilizará una tecnica de remplazo llamada one-hot encoding. Esta tecnica consiste en crear una columna por cada valor nominal, y asignar un 1 si el valor es el mismo que el de la columna, y un 0 si no lo es.

```python
one_hot_encoded = pd.get_dummies(df['Sex'])
one_hot_encoded
```

Notese que hay 3 valores, `F`, `M` y `I`. Por lo tanto, para simplificar el analisis, se procede a crear dos columnas, `F` y `M`. Si el valor es `F` se asigna un 1 a la columna `F` y un 0 a la columna `M`. Si el valor es `M` se asigna un 1 a la columna `M` y un 0 a la columna `F`. Si el valor es `I` se asigna un 0 a la columna `F` y un 0 a la columna `M`. Esto se realiza para evitar la multicolinealidad.

```python
one_hot_encoded = one_hot_encoded.drop('I', axis=1)

df = pd.concat([df, one_hot_encoded], axis=1)

df = df.drop('Sex', axis=1)

df
```

### 6. Histogramas

A continuación se graficaran histogramas para cada una de las variables. Se utilizará la librería Matplotlib para realizar este análisis. La idea es comprender la distribución de los datos.

```python
import matplotlib.pyplot as plt

def create_histogram(data, bins=10, title="Histogram"):
    values, bins, _ = plt.hist(data, bins=bins)
    plt.title(title)
    plt.xlabel("Values")
    plt.ylabel("Frequency")
    plt.xticks(bins)
    plt.grid(True)
    plt.show()
```
